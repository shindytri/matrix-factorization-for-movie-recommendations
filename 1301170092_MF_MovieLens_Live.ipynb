```
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

import pandas as pd
import numpy as np
import heapq

drivePrefix = '/content/gdrive/My Drive/'
driveSuffixUser = '1301170092_MF_Movielens/user_latent_features2.csv'
driveSuffixItem = '1301170092_MF_Movielens/item_latent_features2.csv'

header_user2 = ['k1', 'k2']
header_item2 = ['k1', 'k2']
df_user2 = pd.read_csv(drivePrefix + driveSuffixUser, sep=',', names=header_user2)
df_item2 = pd.read_csv(drivePrefix + driveSuffixItem, sep=',', names=header_item2)
df_user2
```

```
df_item2
```

```
user_features2 = df_user2.to_numpy()
item_features2 = df_item2.to_numpy()

ratings2 = np.dot(user_features2, item_features2.T)

topten_rec2 = []
id = 1
for i in ratings2:
  topten = heapq.nlargest(10, range(len(i)), i.take)
  topten_rec2.append([id, topten[0], topten[1], topten[2], topten[3], topten[4], topten[5], topten[6], topten[7], topten[8], topten[9]])
  id += 1

topten_rec2
```
